```python
import pandas as pd

df = pd.read_csv("cleaned_output.csv")

# Chuyển nhãn về chữ thường để so sánh
df['Label'] = df['Label'].astype(str).str.strip().str.lower()

# Gán nhãn nhị phân: benign = 0, còn lại là 1
df['Label'] = df['Label'].apply(lambda x: 0 if x == 'benign' else 1)

# Lưu kết quả
df.to_csv("binary_output.csv", index=False)

print("Nhãn đã được mã hóa thành nhị phân (benign = 0, attack = 1)")
```

```python
df = pd.read_csv("binary_output.csv")

total_samples = len(df)

label_counts = df['Label'].value_counts()

print(f"Tổng số mẫu: {total_samples}")
print(f"Số mẫu nhãn 0 (Benign): {label_counts.get(0, 0)}")
print(f"Số mẫu nhãn 1 (Attack): {label_counts.get(1, 0)}")
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6, 4))
label_counts.plot(kind='bar', color=['green', 'red'])

# Tùy chỉnh biểu đồ
plt.title("Phân bố nhãn trong dữ liệu (binary classification)")
plt.xlabel("Nhãn")
plt.ylabel("Số mẫu (Triệu mẫu)")
plt.xticks(ticks=[0, 1], labels=["Benign (0)", "Attack (1)"], rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Hiển thị số trên cột
for i, count in enumerate(label_counts):
    plt.text(i, count + total_samples * 0.005, str(count), ha='center')

# Hiển thị biểu đồ
plt.tight_layout()
plt.show()
```
